```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

```python
def result(Data, ratios, learning_rate, classifier, colors, title):
    plt.subplots(nrows = 4, ncols = 3, figsize = (30,30),constrained_layout = True)
    plt.subplots_adjust(left = 0.2, bottom = 0.1, right = 1, top = 1)
    count = 1
    for row, ratio in enumerate(ratios):

        p_pts = Data.get(ratio).get('p_pts')
        p_pts_y = Data.get(ratio).get('p_pts_y')

        n_pts = Data.get(ratio).get('n_pts')
        n_pts_y = Data.get(ratio).get('n_pts_y')

        for col, c in enumerate(learning_rate):

            plt.subplot(4, 3, count, xlabel = 'x', ylabel = 'y', title = str(title) + str(c))
            count +=1

            classifier.C = c

            data_points = np.vstack((p_pts, n_pts))
            outputs = np.vstack((p_pts_y, n_pts_y)) 

            classifier.fit(data_points, outputs)

            plt.scatter(p_pts[:,0], p_pts[:,1], label = "+ve points")
            plt.scatter(n_pts[:,0], n_pts[:,1], label = "-ve points")

            draw_line(classifier.coef_[0], classifier.intercept_[0], 0, 1)

            plt.legend()
```

```python
learning_rate = list([0.001, 1, 100])
```

```python
ratios = [(100, 2), (100, 20), (100, 40), (100, 80)]
Data = dict()
```

```python
for p_pts_ratio, n_pts_ratio in ratios:
    
    p_pts = np.random.normal(loc = 0, scale = 0.05, size = (p_pts_ratio, 2))
    p_pts_y = np.ones((p_pts.shape[0], 1))

    n_pts = np.random.normal(loc = 0.13, scale = 0.02, size = (n_pts_ratio, 2))
    n_pts_y = np.zeros((n_pts.shape[0], 1))
    
    Data.update({
        (p_pts_ratio, n_pts_ratio): {
            'p_pts': p_pts,
            'p_pts_y': p_pts_y,
            'n_pts': n_pts,
            'n_pts_y': n_pts_y
        }
    })
```

```python
svc = SVC(kernel = 'linear')
colors = ['b', 'g', 'r']
```

```python
result(Data, ratios, learning_rate, svc, colors, 'SVM classifier, C: ')
```

# Observation:
### C{0,Infinity}
1. We can see that the hyperparameter (C) impacts the hyperplane separating the positive and negative class. 

2. From the optimization function, we understand that as C tends to infinity, the model will overfit. 

3. On the other hand, as the C moves closer to 0, the model will underfit (high bias). 

4. In my case, since the points are linearly separated, it is better to figure out which bias VS variance trade off that will work best for our model.

5. However, it does not guarantee that it will perfectly classify new unseen data. Therefore, we can hyperpramaeter tuning for the optimal C.
